```python
import numpy as np
from matplotlib import pyplot as plt


data = np.loadtxt('./pict.dat',delimiter=",", dtype=int).reshape(-1,1024)


def init_weights(patterns):
    dim = patterns.shape[1]
    w=np.zeros((dim,dim))
    
    for i in range(dim):
        for j in range(dim):
            s = 0
            for pattern in patterns:
                s += pattern[i]*pattern[j]
            w[i][j]=(1/(patterns.size)) * s
    return w

def calc_energy(pattern):
    print("hej")
    dim = pattern.size
    energy=0
    
    enrg = np.multiply(w, np.multiply.outer(pattern, pattern.T)).sum()
    
    for i in range(dim):
        for j in range(dim):
            energy-= w[i][j]*pattern[i]*pattern[j]
            
    print("Energy",energy)
    print("enrg",enrg)
    
    return energy



def recall(pattern):
    dim = pattern.size
    count = 0
    energyLevels = []
    
    previousPattern = np.zeros(dim)
    while True:
        out = np.zeros(dim)
        for i in range(dim):
            s = 0
            for j in range(dim):
                s += w[i][j]*pattern[j]
                count += 1

            if s >= 0:
                sign=1
            else:
                sign=-1
                    
            out[i]=sign
            
        calc_energy(out)
        
        pattern = out
        
        
        if check_True(pattern):
            print ("It took:",count,"nr of iterations")
            break
            
        elif np.array_equal(pattern,previousPattern):
            print("Local minimum found in iteration:",count)
            break
        
        previousPattern=pattern
    
    plt.plot(x=range(len(energyLevels)),y=energyLevels)
    plt.show()

def check_True(predicted_pattern):
    solved = False
    for i in range(patterns.shape[0]):
        if np.array_equal(patterns[i],predicted_pattern):
            print("True! It matched pattern nr:",i)
            solved = True
    return solved
        
#print(data[0:3,:])

plt.imshow(data[0].reshape(32,32),interpolation="nearest")
plt.show()
    
w = init_weights(data[0:3,:])
patterns = data[0:3,:]
recall(data[0])
#recall(data[10])
#recall(data[2])
                     
```
